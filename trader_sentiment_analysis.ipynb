## High vs Low Leverage

```python
merged['lev_bin'] = pd.qcut(merged['leverage'], 2, labels=['Low','High'])
merged.groupby(['lev_bin','classification'])['Closed PnL'].mean()
```

## Frequent vs Infrequent Traders

```python
trade_counts = merged.groupby('Account').size()
median_trades = trade_counts.median()
merged['trader_type'] = merged['Account'].map(
lambda x: 'Frequent' if trade_counts[x] > median_trades else 'Infrequent')
segment_perf = merged.groupby(['trader_type','classification'])['Closed PnL'].mean()
segment_perf
```

```python
trade_counts.head()
```

## Consistent Winners

```python
account_win_rate = merged.groupby('Account')['win'].mean()
merged['consistency'] = merged['Account'].map(
lambda x: 'Consistent' if account_win_rate[x] > 0.6 else 'Inconsistent')
consistency_perf = merged.groupby(['consistency','classification'])['Closed PnL'].mean()
consistency_perf
```

```python
winrate = merged.groupby('classification')['win'].mean()
# Plot
plt.figure()
winrate.plot(kind='bar')
plt.title("Win Rate by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Win Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Trade Frequency Comparison

```python
trade_frequency = merged.groupby('classification').size()
trade_frequency
trade_frequency.plot(kind='bar', title="Trade Count by Sentiment")
```

## Traders perform best during Greed phases.
Average PnL is noticeably higher during Greed days compared to Fear or Neutral periods. This suggests that traders benefit from stronger directional momentum when the broader market sentiment is positive.

## Neutral markets are the hardest to trade.
Win rate drops significantly during Neutral sentiment. This likely reflects choppy or range-bound conditions where directional conviction is low and trades are less predictable.

## Position sizes increase during emotional markets.
Trade sizes are larger during Extreme Greed and Fear compared to Neutral periods. Traders appear more aggressive when sentiment is strong (positive or negative), but this increase in exposure does not always translate into proportionally higher win rates.

## Part C 

### Lean into Greed, but selectively.
During Greed periods, performance improves overall. Allocating more capital to traders with historically strong win rates during these periods could improve returns.

### Control risk during Fear and Neutral phases.
During Fear periods, traders tend to increase position size, but win rate does not improve meaningfully. A more disciplined approach  such as limiting position size expansion or tightening risk parameters  may help reduce unnecessary drawdowns.
